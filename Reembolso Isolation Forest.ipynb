```python
treino = pd.read_csv(r'C:\Users\tomas\Downloads\Desafio 3\Pedidos de reembolso de executivos.csv',sep=';')
```

```python
treino['receipt_date']=pd.to_datetime(treino['receipt_date'])
```

```python
treino['day'] = treino['receipt_date'].dt.day
treino['month'] = treino['receipt_date'].dt.month
treino['year'] = treino['receipt_date'].dt.year
```

```python
treino.head(1)
```

```python
X = treino.loc[:,('day','month','year','employee_id','receipt_value','bugged_date')]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelBinarizer()
```

```python
X = preprocessing.normalize(X)
```

```python
treino.corr()
```

Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
clf=IsolationForest(n_estimators=100, max_samples='auto', contamination=float(.12), \
                        max_features=1.0, bootstrap=True, n_jobs=-1, random_state=42, verbose=0)
```

```python
clf.fit(X)
pred = clf.predict(X)
treino['anomaly']=pred
outliers=treino.loc[treino['anomaly']== -1]
outlier_index=list(outliers.index)
```
